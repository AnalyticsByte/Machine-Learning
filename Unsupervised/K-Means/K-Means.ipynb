### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
dataset = pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
dataset.head()
```

### Plot the Data

```python
plt.scatter(dataset['Annual Income (k$)'],dataset['Spending Score (1-100)'])
plt.title('Scatter Plot')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

### Select the Features

```python
X = dataset.iloc[:,[3,4]]  #X = dataset.iloc[:,3:5]
```

```python
X.head()
```

### Clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(3)  # creating object of KMeans clustering with 2 clusters
```

```python
kmeans.fit(X) # Fitting X with the object created
```

### Clustering Results

**Checking the predicted clusters i.e which elements belongs to which clusters**

```python
predicted_clusters = kmeans.fit_predict(X)
predicted_clusters
```

**Let's create a sample dataframe with additional coloumn of predicted clusters to see it more clearly**

```python
data_with_predClusters = dataset.copy()
```

```python
data_with_predClusters['Clusters'] = predicted_clusters  # create a new coloumn with name clusters and 
                                                         # values will be equal to predicted_clusters
```

```python
data_with_predClusters
```

**Plotting the scatter plot with the clusters and different colors**

```python
plt.scatter(data_with_predClusters['Annual Income (k$)'],data_with_predClusters['Spending Score (1-100)'],
            c=data_with_predClusters['Clusters'],cmap='rainbow')
plt.title('Scatter Plot')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

### Selecting the number of clusters

**wcss = within cluster sum of squares**

![alt text](WCSS.png "WCSS")

```python
kmeans.inertia_ # this is called WCSS
```

**Finding optimal number of clusters using elbow **

```python
wcss = []
for i in range (1,11):
    kmeans = KMeans(i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range (1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

From the above diagram its clear that there is a huge drop of elbow line at cluster 5

**Now with 5 clusters creating the KMeans clustering**

```python
kmeans = KMeans(5)
kmeans.fit(X)
```

```python
new_predicted_clusters = kmeans.fit_predict(X) # fit_predict() helps us to find which observation belongs to 
                                               # which cluster
new_predicted_clusters
```

```python
new_data_with_predClusters = dataset.copy()
new_data_with_predClusters['New_Clusters'] = new_predicted_clusters
new_data_with_predClusters
```

```python
plt.scatter(new_data_with_predClusters['Annual Income (k$)'],new_data_with_predClusters['Spending Score (1-100)'],
            c=new_data_with_predClusters['New_Clusters'],cmap='rainbow',s=50)
plt.title('Scatter Plot')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
